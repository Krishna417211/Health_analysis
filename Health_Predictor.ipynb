```python
def calculate_bmi(weight_kg, height_m):
    return weight_kg / (height_m ** 2)

def calculate_lean_mass(weight_kg, fat_percentage):
    return weight_kg * (1.0 - fat_percentage / 100.0)

def estimate_calories_burned(weight_kg):
    # Simple estimate: 24 kcal per kg body weight per day
    return weight_kg * 24

def calculate_cal_balance(cal_intake, weight_kg):
    cal_burned = estimate_calories_burned(weight_kg)
    return cal_intake - cal_burned
```

```python
def predict_fat_percentage(weight_kg, height_m, calories_intake, bmi_input, fat_percentage_input_for_lean_mass):
    """
    Inputs:
      - weight_kg: float
      - height_m: float
      - calories_intake: float
      - bmi_input: float (userâ€provides or you could ignore and compute)
      - fat_percentage_input_for_lean_mass: float (if you ask user fat % for computing lean mass; you might instead compute lean mass differently)
    """
    # Compute or override BMI
    bmi_calc = calculate_bmi(weight_kg, height_m)
    # Use user-provided BMI or the computed one? We choose computed
    bmi_used = bmi_calc

    # Compute lean_mass_kg
    lean_mass_kg = calculate_lean_mass(weight_kg, fat_percentage_input_for_lean_mass)
    # Compute cal_balance
    cal_balance = calculate_cal_balance(calories_intake, weight_kg)

    # Build feature array in the same order as X
    sample = np.array([[weight_kg, bmi_used, calories_intake, bmi_calc, cal_balance, lean_mass_kg]])

    # Predict
    pred = model.predict(sample)[0]
    # Return value
    return float(pred)

# --- Step 4: Build Gradio interface ---
interface = gr.Interface(
    fn = predict_fat_percentage,
    inputs = [
        gr.Number(label="Weight (kg)"),
        gr.Number(label="Height (m)"),
        gr.Number(label="Calories Intake (kcal/day)"),
        gr.Number(label="BMI (Optional)"),
        gr.Number(label="Enter your fat percentage prediction")
    ],
    outputs = gr.Number(label="Percentage of fat in your body is:"),
    title = "Body Fat Percentage Predictor",
    description = "Enter weight, height, calorie intake, BMI (optional) and fat percentage to compute lean mass. Then the model predicts fat percentage."
)

# Launch in Colab with shareable link
interface.launch(share=True)
```
